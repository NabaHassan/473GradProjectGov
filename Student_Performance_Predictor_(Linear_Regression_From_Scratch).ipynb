##1. DATASET CREATION

```python
hours_studied = list(range(1, 101))

exam_score = [
   6,  8, 11, 14, 17, 20, 23, 26, 29, 32,
  35, 38, 41, 44, 47, 50, 53, 56, 59, 62,
  64, 66, 68, 70, 72, 74, 76, 78, 80, 82,
  84, 85, 86, 87, 88, 89, 90, 91, 92, 93,
  94, 95, 95, 96, 96, 97, 97, 98, 98, 99,
  99, 99, 98, 99, 99, 98, 99, 99, 98, 99,
  99, 98, 99, 99, 98, 99, 99, 98, 99, 99,
  98, 99, 99, 98, 99, 99, 98, 99, 99, 98,
  99, 99, 98, 99, 99, 98, 99, 99, 98, 99,
  99, 98, 99, 99, 98, 99, 99, 98, 99, 99
]
```

#2. Data Visualization

```python
import matplotlib.pyplot as plt

plt.plot(hours_studied, exam_score, 'ro')
plt.ylabel('Exam Score')
plt.xlabel('Hours Studied')
plt.title('Exam Score vs Hours Studied')
```

# 3.Model Definition

```python
# Model parameters (start with guesses)
w = 0.0
b = 0.0

# Prediction function
def predict(x):
    return w * x + b
```

#4. Loss Function

```python
#Mean Squared Error
def meanSquaredError():
  loss = 0
  for i in range(len(hours_studied)):
    y_predict = predict(hours_studied[i])
    loss += (y_predict - exam_score[i])**2
  return loss/len(exam_score)

```

```python
print(meanSquaredError())
```

#5. Gradient **Calc**

```python
#Gradient of loss w.r.t w and b

def compute_gradients():
  dw = 0
  db = 0
  for i in range(len(hours_studied)):
    y_predict = predict(hours_studied[i])
    dw += hours_studied[i] * (y_predict - exam_score[i])
    db += (y_predict - exam_score[i])
  dw , db = -2/len(exam_score)*dw , -2/len(exam_score)*db
  return dw,db
```

```python
print(compute_gradients())
```

#6. Gradient Descent Loop(Training)

```python
learning_rate = 1e-5 # Reduced learning rate even further

for i in range(1000):
    dw, db = compute_gradients()
    w -= learning_rate * dw
    b -= learning_rate * db

    if i % 100 == 0:
        print(meanSquaredError())
print(f"w is: {w}")
print(f"b is: {b}")
```
